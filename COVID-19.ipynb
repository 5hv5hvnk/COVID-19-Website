```python
import folium
import pandas

hospitals = pandas.read_excel("covid_hp.xlsx", sheet_name=0)

lati = list(hospitals["LAT"])
longi = list(hospitals["LONG"])
name = list(hospitals["NAME"])

html = """ 
<a href="https://www.google.com/search?q=%%22%s%%22" target="_blank">%s</a><br>
"""

Map1 = folium.Map(location = [28.593778, 77.222744], zoom_start = 5)

grp = folium.FeatureGroup(name = "map")

for lat, long, nam in zip(lati, longi, name):
    search = folium.IFrame(html = html % (nam, nam), width=100, height=50)
    grp.add_child(folium.Marker(location = [lat,long], popup = folium.Popup(search), icon = folium.Icon(color="darkblue")))
    
Map1.add_child(grp)
```

```python
Map1.save("Map1.html")
```
